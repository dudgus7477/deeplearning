# 0. 머신러닝과 딥러닝
- 머신러닝은 데이터베이스, 엑셀 등에 담긴 형식이 정해진 데이터를 잘 처리
- 딥러닝은 일정한 형식이 없는 영상, 음성등의 비정형데이터를 잘 처리

### 0-1. 퍼셉트론의 등장
> 퍼셉트론(Perceptron)
- 프랑크 로젠블라트(Frank Rosenblatt)가 1957년에 제안한 초기 형태의 인공 신경망으로 다수의 입력으로부터 하나의 결과를 내보내는 알고리즘
- 입력값을 바탕으로 신호의 출력 여부를 결정한다는 점에서 뉴런과 비슷
- 각 노드의 가중치와 입력치를 곱한 것을 모두 합한 값이 **활성화 함수**에 의해 판단됨
- 초기의 퍼셉트론에서는 그 값이 임계치(보통 0)보다 크면 뉴런이 활성화되고 결과값으로 1을 출력

인공지능 알고리즘이 점차 발전하면서 해결하려는 문제 상황에 적합한 활성화 함수를 개발하려는 노력이 계속되고 있음

### 0-3.인공신경망(NN)에서 딥러닝 모델(DNN)로

#### 0-3-1.인공신경망의 구조
- 입력층, 은닉층, 출력층으로 구성되어있음.
- 각 층은 여러 개의 노드로 구성되어있음.
- 다층으로 구성하면 신경망이 처리할 수 있는 연산의 복잡도가 증가함.

#### 0-3-2.딥러닝 모델
- 심층 신경망(Deep Neural Network, DNN) : 보통 은닉층이 2개 이상인 신경망을 일컫음.
- 은닉층의 개수가 많아질수록 인공신경망이 ‘깊어졌다(deep)’고 부르며, 이렇게 충분히 깊어진 인공신경망을 러닝 모델로 사용하는 머신러닝 패러다임을 바로 딥러닝(Deep Learning)이라고 함.
- 각 층을 구성하는 노드의 개수가 많아질수록 연산의 공간적, 시간적 복잡도가 증가함.
- 노드가 많을수록 복잡한 연산이 가능하지만, 비용이 많이 든다는 단점이 존재함.

### 0-4.딥러닝 모델의 학습

#### 0-4-1.딥러닝 모델의 학습 과정
1. 임의의 가중치 W 설정
2. 순전파를 통해 Cost 계산
######(순전파 : 입력 데이터를 기반으로 신경망을 따라 입력층부터 출력층까지 차례대로 변수들을 계산하고 추론한 결과)
3. 오차역전파를 통하여 Gradient(기울기) 계산
4. W 가중치 조정 $W(t+1) = W(t)-\frac{∂Cost}{∂W} $
5. 2~5의 과정 반복

#### 0-4-2.경사하강법(Gradient Descent)

- 오차 함수의 기울기(경사)를 구하여 기울기가 낮은 쪽으로 계속 이동시켜서 극값에 이를 때까지 반복시키는 것

#### 0-4-3.옵티마이저

- 손실 함수의 값을 줄여나가면서 학습하는 방법에 대한 알고리즘
- 경사하강법(GD)에서 확률적 경사하강법(SGD)이 파생되었으며, 딥러닝 모델을 최적화하기 위한 다양한 알고리즘이 개발 중에 있음
- 안정적인 성능을 보여 현재 대표적으로 많이 사용되는 옵티마이저는 SGD, RMSProp, Adam이 있음

#DNN 무작정 따라하기
- 폴란드의 브로츠와프 의과대학에서 2013년 공개한 폐암 수술 환자의 수술 전 진단 데이터와 수술 후 생존 결과 기록한 실제 의료 데이터를 분석하여 폐암 환자의 생존 여부를 판단한다.
> 17열까지는 의료 기록 데이터 : 종양유형, 폐활량, 호흡 곤란 여부, 고통 정도, 기침, 흡연, 천식 여부 등

```python
# 필요한 라이브러리
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
#실행할 때마다 같은 결과를 출력하기 위해 설정
seed = 0
np.random.seed(seed)
```

```python
#준비된 수술 환자 데이터 불러들이기
df = np.loadtxt("ThoraricSurgery.csv", delimiter=",")
print(df[:3])
```

```python
#환자의 기록과 수술 결과를 X와 Y로 구분하여 저장
X = df[:,0:17]
Y = df[:,17]
```

```python
#딥러닝 구조를 결정
model = Sequential()
#입력층, 은닉층
model.add(Dense(30, input_dim=17, activation='relu')) #퍼셉트론 개수 30, 입력 개수 = 속성 개수(17)
#출력층
model.add(Dense(1, activation='sigmoid')) #분류하고자 하는 개수 0또는 1이면 1
```

image.png

```python
#딥러닝 실행
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy']) #실행할때마다 metrics 정확도 출력
model.fit(X, Y, epochs=30, batch_size = 10) #30번 반복해서 훈련하고 데이터는 10개씩

#결과 출력
print("\n Accuracy : %.4f" %(model.evaluate(X, Y)[1]))
```
